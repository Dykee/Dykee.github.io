```python
@ipywidgets.interact(style = plt.style.available,
                    city = np.unique(climate1["City"]),
                    city2 = np.unique(climate1["City"]))
def make_plot(style, city, city2):
    myCity = climate1[climate1["City"] == city]
    myCity2 = climate1[climate1["City"] == city2]
    with plt.style.context(style):
        #fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
        fig,ax=plt.subplots(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
        p1, = ax.plot(myCity["dt"], myCity["AverageTemperature"], marker='.', linewidth = 1.0, color="pink")
        p2, = ax.plot(myCity2["dt"], myCity2["AverageTemperature"], marker='.', linewidth = 1.0, color="purple")
        ax.set_xlabel("Date", fontsize = 30)
        ax.set_ylabel("Average Temperature", fontsize = 30)
        #ax = plt.gca()
        ax.xaxis.label.set_color('black')
        ax.yaxis.label.set_color('black')
        ax.tick_params(axis = 'both', which = 'major', labelsize = 24, colors = "black")
        ax.tick_params(axis = 'both', which = 'minor', labelsize = 16, colors = "black")
        handles, labels = ax.get_legend_handles_labels() # return lines and labels
        interactive_legend = plugins.InteractiveLegendPlugin(zip(handles, ax.collections),
                                                             labels,
                                                             alpha_unsel=0,
                                                             alpha_over=0, 
                                                             start_visible=True)
        plugins.connect(fig, interactive_legend)
        ax.legend([p1, p2], ["city", "city2"],fontsize = 30)
        # I would like to make the x label and y label visible, dont know how 
        # look up how to change the legend labels to the appropriate city
        # look at bqplot, week 6 in dataviz
```
